```python
import numpy as np
```

Код для анализа эффективности команд роботов-уборщиков. В умном городе две команды роботов работают в разных районах. Представлены их показатели за 6 дней

```python
# Создание массивов с данными о собранном мусоре (в кг). Команда Alpha
alpha_team = np.array([23, 17, 25, 20, 19, 22])
print("Команда Alpha (кг):", alpha_team)
```

```python
# Команда Beta - генерируем прогноз эффективности
beta_team = np.ones(6) * 21
print("Команда Beta (кг):", beta_team)
```

```python
# Сравнение производительности в динамике
days = np.arange(1, 8) # дни недели с понедельника по субботу
print("\nДни наблюдения:", days)
```

```python
# Анализ характеристик массивов
print("Форма массива:", alpha_team.shape) # размерность массива
print("Тип данных:", alpha_team.dtype) # тип элементов
```

```python
# Детальный анализ по дням
print("Результат в среду:", alpha_team[2]) # третий день (индекс 2)
print("Результаты срединной недели:", alpha_team[2:5]) # среда-пятница
```

```python
# Сводная статистика и сравнение команд
total_collected = alpha_team + beta_team
print("Суммарный сбор обеих команд:", total_collected)
```

```python
# Расчёт ключевых метрик
print("\nКлючевые метрики эффективности:")
print("Средний показатель Alpha:", np.mean(alpha_team))
print("Максимальный показатель Beta:", np.max(beta_team))
print("Общий объём уборки:", np.sum(total_collected))
```
